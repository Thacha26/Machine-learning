# 4. Missing Values Analysis

Se identifica, cuantifica y ordena los **datos faltantes** (nulos o *Nan*), para decidir cómo limpiarlos

```python
# Recuento y porcentaje de valores faltantes
missing = pd.DataFrame({ # Nueva tabla llamada missing
    #Número total de datos faltantes
    "Missing Count": df.isnull().sum(), #Calcula la suma total de valores nulos por columna vscíos=true(1), llenos=False(0)
    #Porcentaje de datos faltantes respecto al total de filas
    "Missing %": (df.isnull().sum()/len(df)*100).round(2) # len(df) redodendea a dos decimales
})
        # FIltra la tabla para mostrar solo las columas que tienen un dato faltante
missing = missing[missing["Missing Count"]> 0].sort_values("Missing %", ascending=False)
                                              # Ordena los resultados de mayor a menor procentaje, prioriza las columnas con más problemas de datos
print("Colums with missing values:")
missing
```

#### Porqué hacer esto?
Te dice si:
- **Eliminar** una columna (si tiene >50-70% de faltantes)
- **Rellenar** los datos (usando la media, mediana o el valor más frecuente) si los faltantes son pocos
- **Ignorar** la columna está completa
