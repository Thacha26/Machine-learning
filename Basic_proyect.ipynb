 Predecir el precio de una casa basándote únicamente en su tamaño (en $m^2$)

```python
from sklearn.linear_model import LinearRegression
```

```python
X= [[50], [80], [100], [120], [150]] #metros
y= [100, 150, 190, 220, 280] # precios

modelo = LinearRegression()
modelo.fit(X, y) # se entrena el modelo con la relación entre x e y
nueva_casa = [[110]]
precio_predicho = modelo.predict(nueva_casa)

print(f"El precio estimado para una casa de 110m² es: ${precio_predicho[0]:.2f}k")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([[50], [80], [100], [120], [150]])
y = np.array([100, 150, 190, 220, 280])

model = LinearRegression()
model.fit(X, y)

plt.scatter(X, y, color='blue', label='Datos reales')

X_linea = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_linea = model.predict(X_linea)
plt.plot(X_linea, y_linea, color='red', label='Regresión lineal')

pred_110 = model.predict([[110]])
plt.scatter(110, pred_110, color='green', label='Predicción para 110m²')

plt.xlabel('Superficie (m²)')
plt.ylabel('Precio (miles USD)')
plt.title('Regresión Lineal: Precio vs Tamaño')
plt.legend()
plt.grid(True)

plt.show()
```
